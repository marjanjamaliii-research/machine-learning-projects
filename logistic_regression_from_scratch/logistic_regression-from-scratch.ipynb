# Logistic Regression from Scratch

This project implements logistic regression using only NumPy.

## Goal
Predict binary labels (0 or 1) using a simple model.

## Key Concepts
- Sigmoid function
- Probability
- Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt

#Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#Data (fake simple dataset)
np.random.seed(0)
X = np.random.randn(100, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

#Initialize weights
weights = np.zeros(2)
bias = 0
lr = 0.1

#Training
losses = []

for epoch in range(100):

    linear = np.dot(X, weights) + bias
    preds = sigmoid(linear)

    # loss (binary cross entropy)
    loss = -np.mean(y * np.log(preds + 1e-8) + (1 - y) * np.log(1 - preds + 1e-8))
    losses.append(loss)

    # gradients
    dw = np.dot(X.T, (preds - y)) / len(y)
    db = np.mean(preds - y)

    # update
    weights -= lr * dw
    bias -= lr * db

#Plot loss
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

#Accuracy
predictions = (sigmoid(np.dot(X, weights) + bias) > 0.5).astype(int)
accuracy = np.mean(predictions == y)

print("Accuracy:", accuracy)
```
